```python
spec=[]
n_page_dealer=50        #number of pages that will be visited for each link.
n_page_owner=50
```

```python
def get_Price_owner(site_link):   # Function visits pages, that cars listed by their owners

  for i in range(n_page_owner):
    try:

      i=i+1
      
      if i%5==0:
        print(i)

      pagelink=site_link+str(i)   # writes the page number to be able to visit next pages
      r = requests.get(pagelink)
      soup = BeautifulSoup(r.content, "lxml")
      st1 = soup.find("div", attrs={"class":"flex1 pr"})
      st2 = st1.find("table",attrs= {"class":"table listing-table w100 border-grey2"})
      st3 = st2.find_all("td", attrs={"class":"listing-modelname pr"})

      for ilan in st3:

        try:

          link_tail = ilan.a.get("href")            # gets the unique part of the page link from the listing table.
          link_head = "https://www.arabam.com"
          link=link_head+link_tail                  # combines the standart part and the unique part to get the page link.
          r_1 = requests.get(link)

          soup1 = BeautifulSoup(r_1.content,"lxml")
          details=soup1.find("div", attrs={"class":"banner-column-detail bcd-mid-extended p10 bg-white"})
          
          abs=details.find("div", attrs={"class":"df df-center df-wrap w66 fl"})
          fiyat=int(abs.find("span", attrs={"class":"color-red4 font-semi-big bold fl "}).text.replace(" ","").replace("TL","").replace(".",""))
          
          adress=details.find("p", attrs={"class":"one-line-overflow font-default-minus pt4 color-black2018 bold"}).text.replace(" ","")
          adress=adress.split("/")
          city1=adress[0]
          city2=adress[1]

          car_details = details.find("ul", attrs={"class":"w100 cf mt16"})
          table       = car_details.find_all("li", attrs={"class":"bcd-list-item"})
          ilan_no     = table[0].text.strip().split(":")[1][1:]
          ilan_tarihi = table[1].text.strip().split(":")[1][1:]
          marka       = table[2].text.strip().split(":")[1][1:]
          seri        = table[3].text.strip().split(":")[1][1:]
          model       = table[4].text.strip().split(":")[1][1:]
          yıl         = int(table[5].text.strip().split(":")[1][1:])
          yakıt_tipi  = table[6].text.strip().split(":")[1][1:]
          vites_tipi  = table[7].text.strip().split(":")[1][1:]
          
          if table[8].text.strip().split(":")[1][1:].find("-")>0:
            motor_hacmi = int(table[8].text.strip().split(":")[1][1:].split("-")[1].replace(" cm3",""))
            motor_gucu  = int(table[9].text.strip().split(":")[1][1:].split("-")[1].replace(" HP",""))
          else:
            motor_hacmi = int(table[8].text.strip().split(":")[1][1:].replace(" cc",""))
            motor_gucu  = int(table[9].text.strip().split(":")[1][1:].replace(" hp",""))

          kilometre   = int(table[10].text.strip().split(":")[1][1:].replace(" km","").replace(".",""))
          
          if len(table)>14:
              boya_degişen= table[13].text.strip().split(":")[1]
              kimden      = table[15].text.strip().split(":")[1][1:]
          else:
              boya_degişen= table[11].text.strip().split(":")[1]
              kimden      = table[13].text.strip().split(":")[1][1:]


          subPartDetails=soup1.find("div", attrs={"class":"cf p20"})
          partTable = subPartDetails.find_all("li", attrs={"class":"vertical-half-offset"})
          if len(partTable)>0:
              sag_arka_camurluk= partTable[0].text.strip().split(":")[1][2:]
              arka_kaput        = partTable[1].text.strip().split(":")[1][2:]
              sol_arka_camurluk  =partTable[2].text.strip().split(":")[1][2:]
              sag_arka_kapi     = partTable[3].text.strip().split(":")[1][2:]
              sag_on_kapi       = partTable[4].text.strip().split(":")[1][2:]
              tavan             = partTable[5].text.strip().split(":")[1][2:]
              sol_arka_kapi     = partTable[6].text.strip().split(":")[1][2:]
              sol_on_kapi       = partTable[7].text.strip().split(":")[1][2:]
              sag_on_camurluk   = partTable[8].text.strip().split(":")[1][2:]
              motor_kaputu      = partTable[9].text.strip().split(":")[1][2:]
              sol_on_camurluk   = partTable[10].text.strip().split(":")[1][2:]
              on_tampon         = partTable[11].text.strip().split(":")[1][2:]
              arka_tampon       = partTable[12].text.strip().split(":")[1][2:]
          else:
              sag_arka_camurluk = "n/a"
              arka_kaput        = "n/a"
              sol_arka_camurluk = "n/a"
              sag_arka_kapi     = "n/a"
              sag_on_kapi       = "n/a"
              tavan             = "n/a"
              sol_arka_kapi     = "n/a"
              sol_on_kapi       = "n/a"
              sag_on_camurluk   = "n/a"
              motor_kaputu      = "n/a"
              sol_on_camurluk   = "n/a"
              on_tampon         = "n/a"
              arka_tampon       = "n/a"

          spec.append([fiyat,city1,city2,ilan_no,ilan_tarihi,marka,seri,model,yıl,yakıt_tipi,vites_tipi,motor_hacmi,motor_gucu,kilometre,boya_degişen,kimden,\
                      sag_arka_camurluk,arka_kaput,sol_arka_camurluk,sag_arka_kapi,sag_on_kapi,tavan,sol_arka_kapi,sol_on_kapi,sag_on_camurluk,motor_kaputu,sol_on_camurluk,on_tampon,arka_tampon,link
                      ])


        except:
          pass
    except:
      break

```
